#### Chart - 10 Distance from Prime Location to Availability by Room Type

```python
# Define the coordinates of the prime location (e.g., Times Square)
prime_location = (40.7580, -73.9855)  # (latitude, longitude)

# Haversine formula to calculate distance
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers
    lat1_rad, lon1_rad = np.radians(lat1), np.radians(lon1)
    lat2_rad, lon2_rad = np.radians(lat2), np.radians(lon2)

    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    return distance

# Calculate the distance for each listing
Adf['distance'] = Adf.apply(lambda row: haversine(row['latitude'], row['longitude'], prime_location[0], prime_location[1]), axis=1)
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=Adf, x='distance', y='availability_365', hue='room_type', style='room_type', palette='deep')
plt.title('Distance from Prime Location to Availability by Room Type')
plt.xlabel('Distance from Prime Location (in miles/kilometers)')
plt.ylabel('Availability (Days per Year)')
plt.legend(title='Room Type')
plt.show()
```
